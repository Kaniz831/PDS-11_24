```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1
url = "https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv"
df = pd.read_csv(url)

# Step 2
null_columns = df.columns[df.isnull().any()]
print("Columns with Null values:", null_columns)

# Step 3
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Cabin'].fillna('Unknown', inplace=True)

# Step 4
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Sex', data=df, palette='Set2')
plt.xticks([0, 1], ['Dead', 'Survived'])
plt.title("Survival Count by Gender")
plt.show()

# Step 5
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Pclass', data=df, palette='Set1')
plt.xticks([0, 1], ['Dead', 'Survived'])
plt.title("Survival Count by Passenger Class")
plt.show()

# Step 6
df['AgeClass'] = df['Age'].apply(lambda x: 0 if x <= 16 else (1 if x <= 26 else (2 if x <= 36 else (3 if x <= 62 else 4))))

# Step 7
df.drop(columns=['Age'], inplace=True)

print(df.head())
#step 8
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='AgeClass', data=df, palette='viridis')
plt.xticks([0, 1], ['Dead', 'Survived'])
plt.title("Survival Count by Age Class")
plt.xlabel("Survival Status")
plt.ylabel("Count")
plt.legend(title="Age Class")
plt.show()
```
